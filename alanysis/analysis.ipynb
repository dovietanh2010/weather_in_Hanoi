```python
import pandas as pd
from sqlalchemy import create_engine

# Cấu hình database
DB_CONFIG = {
    "dbname": "airflow",
    "user": "airflow",
    "password": "airflow",
    "host": "localhost",
    "port": 5432
}

# Tạo kết nối
engine = create_engine(f"postgresql://{DB_CONFIG['user']}:{DB_CONFIG['password']}@{DB_CONFIG['host']}:{DB_CONFIG['port']}/{DB_CONFIG['dbname']}")

# Đọc dữ liệu từ PostgreSQL
query = "SELECT * FROM weather_data"
df = pd.read_sql(query, engine)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(df['timestamp'], df['temperature'], marker='o', linestyle='-', label='Temperature (°C)', color='red')
plt.plot(df['timestamp'], df['humidity'], marker='o', linestyle='-', label='Humidity (%)', color='blue')
plt.plot(df['timestamp'], df['wind_speed'], marker='o', linestyle='-', label='Wind Speed (m/s)', color='green')

plt.xlabel("Timestamp")
plt.ylabel("Value")
plt.title("Weather Data Trends Over Time")
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```
